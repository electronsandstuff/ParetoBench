# CF Pareto Fronts
This notebook plots some of the Pareto fronts frmo the CF problems to confirm they appear correct.

```python
import matplotlib.pyplot as plt
import paretobench as pb
```

```python
f = pb.CF8().get_pareto_front(5024)
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.scatter(f[:, 0], f[:, 1], f[:, 2])
ax.view_init(elev=30, azim=30)
```

```python
f = pb.CF10().get_pareto_front(5024)
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.scatter(f[:, 0], f[:, 1], f[:, 2])
ax.view_init(elev=30, azim=30)
```
