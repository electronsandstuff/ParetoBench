```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

```python
# Since the second objective just depends on g and the first, we need to minimize the first objective
def zdt6_obj1(x):
    return 1 - np.exp(-4 * x) * (np.sin(6 * np.pi * x)) ** 6


# Use a numerical method to find the minimum
x = np.linspace(0, 1, 2048)
y = zdt6_obj1(x)
guess = x[np.argmin(y)]
minimum = minimize(zdt6_obj1, guess, bounds=[[guess * 0.9, guess * 1.1]], tol=1e-13).x[0]

# Plot everything
x = np.linspace(0, 1, 256)
y = zdt6_obj1(x)
plt.plot(x, y)
plt.plot(minimum, zdt6_obj1(minimum), marker="o", ls="none")
plt.xlabel("Decision Var 1")
plt.ylabel("Objective 1")
zdt6_obj1(minimum)
```
